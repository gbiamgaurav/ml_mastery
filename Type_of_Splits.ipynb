## Train Test Split 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
features = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

```python
df = pd.read_csv(filename, names=features)
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
display(df.to_string())
```

```python
array = df.values
```

```python
X = array[:, 0:8]
y = array[:, 8]
```

```python
X
```

```python
y
```

```python
test_size=0.33
random_state = 7
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
result = model.score(X_test,y_test)
```

```python
print("Accuracy : ", result*100)
```

## K Fold Cross Validation

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
import warnings
warnings.filterwarnings("ignore")
```

```python
kfold = KFold(n_splits=10)
model = LogisticRegression()
results = cross_val_score(model, X, y, cv=kfold)
print("Accuracy :", results.mean()*100, results.std()*100)
```

## Leave One Out Cross Validation

```python
from sklearn.model_selection import LeaveOneOut
from sklearn.model_selection import cross_val_score
```

```python
loocv = LeaveOneOut()
model = LogisticRegression()
results = cross_val_score(model, X, y, cv=loocv)
print("Accuracy :", results.mean()*100, results.std()*100)
```

## Repeated Random Train Test Splits 

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
kfold = ShuffleSplit(n_splits=10, test_size=test_size)
model = LogisticRegression()
results = cross_val_score(model, X, y, cv=kfold)
print("Accuracy :", results.mean()*100, results.std()*100)
```
